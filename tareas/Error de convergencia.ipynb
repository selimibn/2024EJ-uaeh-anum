```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

Vamos a definir una funci√≥n recursiva

```python
N = 50
y = np.empty(N+1)            # Allocate an empty vector with N+1 entries

# Now use the difference equation to generate the numbers in the sequence
y[0] = 1
y[1] = 1/5
for z in range(2,N+1):
    y[z] = 16/5*y[z-1] - 3/5*y[z-2]

n = np.arange(N+1)

plt.semilogy(n,y, 'rx', markersize=5, label='$y_n$')
plt.semilogy(n,(1/5)**n, 'bx', markersize=5, label='$y_n$')
plt.show()
```
